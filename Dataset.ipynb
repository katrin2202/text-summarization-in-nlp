````python
dataset3_val_text = dataset3['validation']['text']
print(dataset3_val_text[0])
dataset3_val_summ = dataset3['validation']['summary']
print(dataset3_val_summ[0])
````

````python
dataset3_test_text = dataset3['test']['text']
print(dataset3_test_text[0])
dataset3_test_summ = dataset3['test']['summary']
print(dataset3_test_summ[0])
````

````python
import random

x_train = []
y_train = []

x_train[:] = dataset1_train_text[:]
print(len(x_train))
x_train[len(x_train):] = dataset2_train_text[:]
print(len(x_train))
x_train[len(x_train):] = dataset3_train_text[:]
print(len(x_train))

y_train[:] = dataset1_train_summ[:]
print(len(y_train))
y_train[len(y_train):] = dataset2_train_summ[:]
print(len(y_train))
y_train[len(y_train):] = dataset3_train_summ[:]
print(len(y_train))
````

````python
x_val = []
y_val = []

x_val[:] = dataset1_val_text[:]
print(len(x_val))
x_val[len(x_val):] = dataset2_val_text[:]
print(len(x_val))
x_val[len(x_val):] = dataset3_val_text[:]
print(len(x_val))

y_val[:] = dataset1_val_summ[:]
print(len(y_val))
y_val[len(y_val):] = dataset2_val_summ[:]
print(len(y_val))
y_val[len(y_val):] = dataset3_val_summ[:]
print(len(y_val))
````

````python
x_test = []
y_test = []

x_test[:] = dataset1_test_text[:]
print(len(x_test))
x_test[len(x_test):] = dataset2_test_text[:]
print(len(x_test))
x_test[len(x_test):] = dataset3_test_text[:]
print(len(x_test))

y_test[:] = dataset1_test_summ[:]
print(len(y_test))
y_test[len(y_test):] = dataset2_test_summ[:]
print(len(y_test))
y_test[len(y_test):] = dataset3_test_summ[:]
print(len(y_test))
````

````python
import random

random.shuffle(x_train) 
random.shuffle(y_train) 
random.shuffle(x_val) 
random.shuffle(y_val) 
random.shuffle(x_test) 
random.shuffle(y_test) 
````

````python
from datasets.dataset_dict import DatasetDict
from datasets import Dataset

dataset = {'train':Dataset.from_dict({'summ':y_train,'text':x_train}),
     'val':Dataset.from_dict({'summ':y_val,'text':x_val}),
     'test':Dataset.from_dict({'summ':y_test,'text':x_test})
     }

DatasetDict(dataset)
````

````python

# assume that we have already loaded the dataset called "dataset"
for split, data in dataset.items():
  print(split)
  print(data)
  data.to_csv(f"my-dataset-{split}.csv", index = None)
````
